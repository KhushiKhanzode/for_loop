```python
#REMOVE DUPLICATE CHARACTER FROM STRING
a = "engineering"
b = ''
for i in a:
     if i not in b:
          b+=i
print(b)

```

```python
#PRINT * IN PLACE OF VOWELS IN STRING
c = "Developer"
for i in c:
    if i == 'a' or i == 'e' or i == 'i' or i =='o' or i == 'u':
        b = i.replace(i,"*")
        print (b,end='')
    else:
        print(i,end='')
    
```

```python
# PRINT ONLY EVEN PLACE CHARACTER
a = 'ENGINEERING'
for i in range(1,len(a),2):
        print(a[i],end = ' ')
```

```python
#COUNT VOWELS IN GIVEN STRING
a = 'education'
count = 0
for i in  a:
    if i == 'a' or i == 'e' or i == 'i' or i =='o' or i == 'u':
        count+=1
print(f"Total count of vowels in given string is {count}")
```

```python
# PRINT LONGEST WORD IN SENTENCE
a = "Data Science and Machine learning"
b = a.split(" ")
g=b[0] 

for i in b:
    if len(g) <= len(i):
        g =i

print(f"longest word is {g}")


```
